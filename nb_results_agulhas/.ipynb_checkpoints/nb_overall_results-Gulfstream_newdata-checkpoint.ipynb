## <u> 3. Lagrangian Cumulative Distance</u> (at 15m)

```python
dir_drifter = '../dc_data/drifter/'

dir_advection = '../dc_data/advection/'

drifter_list = [dir_drifter + 'Drifters_AOML_region_T1_15m_20190101T000000Z_20200101T000000Z.pyo.gz']

all_path_artificial_drifters = [dir_advection + 'advection_T1_008_047_15m/data_type_008_047_region_T1_dep1.pyo.gz',
                                dir_advection + 'advection_T1_cmems_015_004_15m/data_type_cmems_015_004_region_T1_dep1.pyo.gz',
                                dir_advection + 'advection_T1_woc-l4-curgeo-bfn-3h_15m/data_type_woc-l4-curgeo-bfn-3h_region_T1_dep1.pyo.gz',
                                dir_advection + 'advection_T1_woc-l4-cureul-glob-1h_15m/data_type_woc-l4-cureul-glob-1h_region_T1_dep1.pyo.gz']

all_outputfilename = ['SDE_region_T1_008_047_15m',
                      'SDE_region_T1_015_004_15m',
                      'SDE_region_T1_woc-l4-curgeo-bfn-3h_15m',
                      'SDE_region_T1_woc-l4-cureul-glob-1h_15m']
```

```python


for path_artificial_drifters, outputfilename in zip(all_path_artificial_drifters, all_outputfilename):
    
    print(outputfilename)

    _ = sde.run(path_artificial_drifters, drifter_list, output_dir=outputdir, output_filename=outputfilename)
```

```python
list_sde = [outputdir + output_filename + '.pyo.gz' for output_filename in all_outputfilename]

outputfilename = 'SDE_region_T1_15m.png'

_ = sde.plot(list_sde, output_dir=outputdir, 
             output_filename=outputfilename,
             list_color=['r','b','c','g'])
```
